```python
response = driver.page_source
soup = bs(response)

# 서울시 전체 개수 : 검색결과는 580으로 나오는데 590이 나오는 것은 지도에서는 10을 추가하기 때문이다.
# 스타벅스 사이트에서의 서울 지역 매장 개수는 580개가 있지만 
# 일어온 결과는 약간의 차이가 있다. 580 -> 590  현재 위치의 주변에 있는 매장 10개를 한번 더 카운트 하였기 때문에 차이가 있다.
# 따라서 뒤에서 중복 값을 제거한다. 

len(soup.select('li.quickResultLstCon'))
```

```python
stars = soup.select('li.quickResultLstCon')
star = stars[0]
print(star)
```

```python
# 추출해야 할 정보 4 가지
# 위도
lat = star['data-lat']   # 키값 밸류값 형태의 딕셔너리 추출
print(lat)

# 경도
lng = star['data-long']
print(lng)

# 주소
# addr = star.select('p.result_details')[0].text
addr = str(star.select('p.result_details')[0]).split('<br/>')[0].split('>')[1]  # br을 기준으로 자르자.-> 그런데 자를 수 없다. 그래서 전체를 str로 바꾼다
print(addr)

# 전화번호
tel = str(star.select('p.result_details')[0]).split('<br/>')[1].split('<')[0]  
tel

# 지점명
name = star['data-name']
print(name)
```

```python
# 전체 지점 데이터 프레임 만들기
import pandas as pd

starbucks_list = []

stars = soup.select('li.quickResultLstCon')


for star in stars:
  
  name = star['data-name']
  lat = star['data-lat']   
  lng = star['data-long']
  addr = str(star.select('p.result_details')[0]).split('<br/>')[0].split('>')[1] 
  tel = str(star.select('p.result_details')[0]).split('<br/>')[1].split('<')[0]   

  starbucks_list.append([name,lat,lng,addr,tel])

```

```python
  df = pd.DataFrame(starbucks_list, columns=['지점명','위도','경도','주소','전화번호'])
  df
# df.to_excel('movie.xlsx', index=False)
```

```python
# 데이터 프레임에서 중복 항목 제거
df2 = df.drop_duplicates()
df2
```

```python
# 서울시 구별 스타벅스 매장수 파악

# df2 에 '시군구' 라는 열 이름을 생성하여 '주소' 열에서 구만 추출하여 넣는다. 

sgg_names = []                 # 구의 이름만 추출하여 담을 리스트 생성

for addr in df2['주소']:       # 주소를 하나씩 추출
  sgg = addr.split()[1]        # 빈판으로 분리하여 [1]을 전부 추출
  sgg_names.append(sgg)        # 구의 이름들만 리스트에 담기

df2['시군구명'] = sgg_names    # 구 이름 리스트를 데이터 프레임에 그대로 삽입
df2
```

```python
# 구별 스타벅스의 매장수 파악
# a = df2.groupby(['시군구명']).count()
# print(a)

# 좀 더 보기 좋게 표현
star_count = df2.pivot_table(index = '시군구명',   # 구분할 항목   
                             values = '지점명',    # 계산할 열
                             aggfunc = 'count'     
                             ).rename(columns={'지점명':'지점수'})
star_count
```

```python
# 서울시 구별 인구수, 종사자수 파일 읽어오기

seoul_sgg = pd.read_excel('sgg_pop.xlsx', thousands=',')   # 서울시 구별 인구수, 천단위 구분기호 ,삭제
seoul_biz = pd.read_excel('sgg_biz.xlsx')   # 서울시 구별 종사자수
# seoul_sgg
# seoul_biz

# star_count                                 # 서울시 구별 스타벅스 매장 수
```

```python
# 세 개의 데이터 프레임을 하나로 병합
# 각각 2개씩 순차적으로 병합한다. 

# 먼저 서울시 구별 인구수와 구별 스타벅스 매장수를 병합
seoul_sgg = pd.merge(seoul_sgg, star_count, how='left', on='시군구명')
seoul_sgg
```

```python
# 다시 근무자수 병합

seoul_sgg = pd.merge(seoul_sgg, seoul_biz, how='left', on='시군구명') 
seoul_sgg
```

```python
# 주민등록인구 십만명당 매장수

seoul_sgg['거주자_십만명당_매장수'] = seoul_sgg['지점수'] / (seoul_sgg['주민등록인구'] / 100000)
seoul_sgg
```

```python
# 종사자수 십만명당 매장수
seoul_sgg['종사자수_십만명당_매장수'] = seoul_sgg['지점수'] / (seoul_sgg['종사자수'] / 100000)
seoul_sgg
```

```python
# 사업체수 만명당 매장수
seoul_sgg['업체수_만명당_매장수'] = seoul_sgg['지점수'] / (seoul_sgg['사업체수'] / 10000)
seoul_sgg
```
