```python
# 시리즈와 데이터프레임에 대해 자세히 알아보자

# 시리즈 만들기

import pandas as pd

a=pd.Series(['apple','bnanana','cherry'])
print(type(a))
print(a)

a2 = pd.Series(['apple','bnanana','cherry'],index=['a','b','c'])
print(a2)
```

```python
# 데이터프레임 만들기

person = pd.DataFrame({
    'name':['leekh','kelly'],
    'job':['백수','녹고먹기'],
    'born':['1898-08-24','1700-03-23'],
    'age' :[44,21]
})

print(person)
print(type(person))

# 시리즈 처럼 데이터프레임도 인덱스를 따로 지정하지 않으면 인덱스를 0부터 자동으로 생성한다. 
```

```python

# 인덱스 값 직접 지정
person = pd.DataFrame({
    'name':['leekh','kelly'],
    'job':['백수','녹고먹기'],
    'born':['1898-08-24','1700-03-23'],
    'family':[5,4],
    'age' :[44,21] 
},index=['first','second'])

print(person)
print(type(person))

# 딕셔너리는 순서를 보장하지 않는다. 따라서 순서까지 보장된 딕셔너리를 전달하려면 
```

```python
from collections import OrderedDict
# 인덱스 값 직접 지정
# 딕셔너리는 순서를 보장하지 않는다. 따라서 순서까지 보장된 딕셔너리를 전달하려면 
# OrderedDict 모듈을 사용한다. 
# 딕셔너리의 데이터 순서를 그대로 유지하면서 데이터프레임을 만들 수 있다. 

person = pd.DataFrame(OrderedDict({
    'name':['leekh','kelly'],
    'job':['백수','녹고먹기'],
    'born':['1898-08-24','1700-03-23'],
    'family':[5,4],
    'age' :[44,21] 
})
)
print(person)
print(type(person))
```

```python
# 판다스의 데이터 구성의 가장 기본은 시리즈이다. 

# 데이터프레임에서 시리즈를 선택하려면 loc 속성에 인덱스를 전달한다. 

person = pd.DataFrame({
    'name':['leekh','kelly'],
    'job':['백수','녹고먹기'],
    'born':['1898-08-24','1700-03-23'],
    'family':[5,4],
    'age' :[44,21] 
},index=['first','second'])

first_one = person.loc['first']  # loc는 인덱스 값으로 선택한다. 
type(first_one)
```

```python
# Series 에서 많이 사용되는 속성과 메소드 정리

person = pd.DataFrame({
    'name':['leekh','kelly'],
    'job':['백수','녹고먹기'],
    'born':['1898-08-24','1700-03-23'],
    'family':[5,4],
    'age' :[44,21] 
},index=['first','second'])

first_one = person.loc['first'] 

# 1. index 속성 - Series의 인덱스 값들만 추출
print(first_one.index)   # index는 속성 추출

print(first_one.values)

print(first_one.keys())   # 인덱스와 동일한 결과를 보여주지만, keys()는 메소드이다. 
print(first_one.keys()[0])
```

```python
# 계산 관련 메소드들

age = person['age']
print(age.mean())  # 평균
print(age.min())   # 최소값
print(age.max())   # 최대값
print(age.std())   # 표준편차
```
